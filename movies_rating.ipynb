```python
import matplotlib.pyplot as plt

# Assuming 'movies' DataFrame is already defined and has the columns 'Genre' and 'BudgetMillions'

# Extract budgets for each genre
action_budgets = movies[movies.Genre == 'Action'].BudgetMillion
drama_budgets = movies[movies.Genre == 'Drama'].BudgetMillion
thriller_budgets = movies[movies.Genre == 'Thriller'].BudgetMillion

# Plot the histogram
plt.hist([action_budgets, drama_budgets, thriller_budgets], bins=20, stacked=True, label=['Action', 'Drama', 'Thriller'])
plt.legend()
plt.xlabel('Budget in Millions')
plt.ylabel('Frequency')
plt.title('Distribution of Budgets by Genre')
plt.show()
```
